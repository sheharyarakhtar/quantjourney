```python
import backtrader as bt
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 1. Download Historical Data using yfinance
symbol = "AAPL"  # Change to any stock symbol
data = yf.download(symbol, start="2020-01-01", end="2020-12-01", multi_level_index=False)

# Convert to Backtrader Format
data_bt = bt.feeds.PandasData(dataname=data)

# 2. Define a Simple Moving Average Crossover Strategy
class MovingAverageCrossStrategy(bt.Strategy):
    def __init__(self):
        self.sma_short = bt.indicators.SimpleMovingAverage(period=10)  # Short SMA
        self.sma_long = bt.indicators.SimpleMovingAverage(period=50)   # Long SMA

    def next(self):
        if self.position:
            if self.sma_short[0] < self.sma_long[0]:  # Sell Condition
                self.close()
        else:
            if self.sma_short[0] > self.sma_long[0]:  # Buy Condition
                self.buy()
    def stop(self):
      final_value = self.broker.get_value()
      print(f"Final Portfolio Value: ${final_value:.2f}")


# 3. Set up Backtrader (Cerebro)
cerebro = bt.Cerebro()
cerebro.addstrategy(MovingAverageCrossStrategy)
cerebro.adddata(data_bt)  # Use Yahoo Finance Data
cerebro.broker.set_cash(10000)  # Set Starting Capital

# 4. Run Backtest & Plot Results
cerebro.run()
from IPython.display import Image

# Set the path to save the image
img_path = "backtrader_plot.png"

# Generate the plot and save it
fig = cerebro.plot(style="candlestick")[0][0]
fig.savefig(img_path)

# Display the saved image
Image(filename=img_path)
```

```python
import numpy as np
import backtrader as bt
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Get stock data
ticker = "AAPL"
df = yf.download(ticker, start="2023-01-01", end="2023-12-01", multi_level_index=False)

# Simple strategy class
class SimpleMACrossStrategy(bt.Strategy):
    def __init__(self):
        self.sma_short = bt.indicators.SimpleMovingAverage(self.data.close, period=10)
        self.sma_long = bt.indicators.SimpleMovingAverage(self.data.close, period=50)
        
        # Track portfolio values and signals
        self.starting_cash = self.broker.get_cash()
        self.portfolio_values = [self.starting_cash]
        self.buy_signals = []
        self.sell_signals = []
        self.holdings = [0]  # Start with cash (0 = cash, 1 = invested)
        self.invested = False
        self.shares = 0
        self.entry_price = 0
        
    def next(self):
        # Calculate current portfolio value based on position
        if self.invested:
            # When invested, value = (current price * shares)
            current_value = self.data.close[0] * self.shares
        else:
            # When in cash, value = starting cash
            current_value = self.starting_cash
            
        self.portfolio_values.append(current_value)
        self.holdings.append(1 if self.invested else 0)
        
        # Check for buy/sell conditions
        if not self.invested:  # If we're in cash
            if self.sma_short[0] > self.sma_long[0]:
                # Buy all-in with cash
                self.shares = self.starting_cash / self.data.close[0]
                self.entry_price = self.data.close[0]
                self.invested = True
                self.buy_signals.append((len(self.data)-1, self.data.close[0]))
                
        else:  # If we're holding stock
            if self.sma_short[0] < self.sma_long[0]:
                # Sell everything - cash becomes shares * current price
                self.starting_cash = self.shares * self.data.close[0]
                self.shares = 0
                self.invested = False
                self.sell_signals.append((len(self.data)-1, self.data.close[0]))


# Set up and run backtest
cerebro = bt.Cerebro()
cerebro.broker.set_cash(10000)  # Starting with $10,000

# Add data feed
data = bt.feeds.PandasData(dataname=df)
cerebro.adddata(data)

# Add strategy
cerebro.addstrategy(SimpleMACrossStrategy)

# Run backtest
results = cerebro.run()
strategy = results[0]

# Prepare data for visualization
dates = df.index.tolist()
prices = df['Close'].values
sma_short = df['Close'].rolling(window=10).mean().values
sma_long = df['Close'].rolling(window=50).mean().values

# Create animation
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), gridspec_kw={'height_ratios': [3, 1]})

# Price chart setup (top subplot)
ax1.set_title(f"{ticker} Price with MA Cross Strategy")
ax1.set_ylabel("Stock Price ($)")
price_line, = ax1.plot([], [], 'k-', label="Stock Price")
sma_short_line, = ax1.plot([], [], 'b-', label="10-day MA")
sma_long_line, = ax1.plot([], [], 'r-', label="50-day MA")
buy_scatter = ax1.scatter([], [], color='green', s=100, label="Buy Signal")
sell_scatter = ax1.scatter([], [], color='red', s=100, label="Sell Signal")
ax1.legend(loc="upper left")
ax1.grid(True)

# Portfolio value chart (bottom subplot)
ax2.set_xlabel("Date")
ax2.set_ylabel("Portfolio Value ($)")
portfolio_line, = ax2.plot([], [], 'g-', label="Portfolio Value")
ax2.legend(loc="upper left")
ax2.grid(True)

# Status text
status_text = ax1.text(0.02, 0.05, '', transform=ax1.transAxes, 
                       fontsize=12, bbox=dict(facecolor='white', alpha=0.7))

# Set axis limits
ax1.set_xlim(0, len(dates))
ax1.set_ylim(min(prices) * 0.95, max(prices) * 1.05)
ax2.set_xlim(0, len(dates))
ax2.set_ylim(min(strategy.portfolio_values) * 0.95, max(strategy.portfolio_values) * 1.05)

def init():
    price_line.set_data([], [])
    sma_short_line.set_data([], [])
    sma_long_line.set_data([], [])
    portfolio_line.set_data([], [])
    status_text.set_text('')
    return price_line, sma_short_line, sma_long_line, portfolio_line, status_text

def update(frame):
    # Update price chart lines
    x = range(frame)
    price_line.set_data(x, prices[:frame])
    
    if frame > 10:
        sma_short_line.set_data(range(10, frame), sma_short[10:frame])
    if frame > 50:
        sma_long_line.set_data(range(50, frame), sma_long[50:frame])
    
    # Update buy/sell signals
    buy_x = [p[0] for p in strategy.buy_signals if p[0] < frame]
    buy_y = [p[1] for p in strategy.buy_signals if p[0] < frame]
    sell_x = [p[0] for p in strategy.sell_signals if p[0] < frame]
    sell_y = [p[1] for p in strategy.sell_signals if p[0] < frame]
    
    # Fix for scatter plot coordinates
    if buy_x:
        buy_scatter.set_offsets(np.column_stack([buy_x, buy_y]))
    else:
        buy_scatter.set_offsets(np.zeros((0, 2)))
        
    if sell_x:
        sell_scatter.set_offsets(np.column_stack([sell_x, sell_y]))
    else:
        sell_scatter.set_offsets(np.zeros((0, 2)))
    
    # Update portfolio value
    portfolio_line.set_data(range(frame), strategy.portfolio_values[:frame])
    
    # Update status text
    if frame < len(strategy.holdings):
        status = "HOLDING STOCK" if strategy.holdings[frame] == 1 else "HOLDING CASH"
        value = f"${strategy.portfolio_values[frame]:.2f}"
        status_text.set_text(f"Status: {status}\nValue: {value}")
    
    return price_line, sma_short_line, sma_long_line, portfolio_line, buy_scatter, sell_scatter, status_text


ani = animation.FuncAnimation(fig, update, frames=len(dates), init_func=init, 
                              interval=50, blit=True)

HTML(ani.to_jshtml())
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

ticker = "BTC-USD"
df = yf.download(ticker, start="2024-01-01", end="2024-12-31", multi_level_index=False)
df['SMA10'] = df['Close'].rolling(window=10).mean()
df['SMA50'] = df['Close'].rolling(window=50).mean()

# Create buy/sell signals
df.dropna(inplace=True)

df['Signal'] = 0
df.loc[df['SMA10'] > df['SMA50'], 'Signal'] = 1  # Buy signal (SMA10 > SMA50)

df['Signal_Change'] = df['Signal'].diff()
df.dropna(inplace=True)
buy_points = df[df['Signal_Change'] == 1].index
sell_points = df[df['Signal_Change'] == -1].index
print(buy_points)
print(sell_points)
initial_cash = 100
df['Portfolio'] = initial_cash  # Start with cash
in_market = False
shares = 0
cash = initial_cash
```

```python


# Calculate portfolio value (starts at 10000 cash)

# Simulate portfolio changes

for i in range(len(df)):
    current_date = df.index[i]
    
    # Buy condition
    if not in_market and current_date in buy_points:
        shares = cash / df.iloc[i]['Close']
        in_market = True
        
    # Sell condition
    if in_market and current_date in sell_points:
        cash = shares * df.iloc[i]['Close']
        shares = 0
        in_market = False
    
    # Update portfolio value
    if in_market:
        df.iloc[i, df.columns.get_loc('Portfolio')] = shares * df.iloc[i]['Close']
    else:
        df.iloc[i, df.columns.get_loc('Portfolio')] = cash

# Get data for animation
dates = df.index
prices = df['Close'].values
sma10 = df['SMA10'].values
sma50 = df['SMA50'].values
portfolio = df['Portfolio'].values

# Create animation
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), height_ratios=[3, 1])

# Price chart setup
ax1.set_title(f"{ticker} Price with MA Cross Strategy")
ax1.set_ylabel("Stock Price ($)")
price_line, = ax1.plot([], [], 'k-', label="Stock Price")
sma10_line, = ax1.plot([], [], 'b-', label="10-day MA")
sma50_line, = ax1.plot([], [], 'r-', label="50-day MA")
buy_scatter = ax1.scatter([], [], color='green', s=100, label="Buy Signal")
sell_scatter = ax1.scatter([], [], color='red', s=100, label="Sell Signal")
ax1.legend(loc="upper left")
ax1.grid(True)

# Portfolio value chart
ax2.set_xlabel("Date")
ax2.set_ylabel("Portfolio Value ($)")
portfolio_line, = ax2.plot([], [], 'g-', label="Portfolio Value")
ax2.legend(loc="upper left")
ax2.grid(True)

# Status text
status_text = ax1.text(0.02, 0.05, '', transform=ax1.transAxes, 
                       fontsize=12, bbox=dict(facecolor='white', alpha=0.7))

# Set axis limits
date_indices = np.arange(len(dates))
ax1.set_xlim(0, len(dates))
ax1.set_ylim(min(prices) * 0.95, max(prices) * 1.05)
ax2.set_xlim(0, len(dates))
ax2.set_ylim(min(portfolio) * 0.95, max(portfolio) * 1.05)

def init():
    price_line.set_data([], [])
    sma10_line.set_data([], [])
    sma50_line.set_data([], [])
    portfolio_line.set_data([], [])
    status_text.set_text('')
    return price_line, sma10_line, sma50_line, portfolio_line, status_text

def update(frame):
    # Update price chart
    price_line.set_data(date_indices[:frame], prices[:frame])
    
    if frame > 10:
        sma10_line.set_data(date_indices[9:frame], sma10[9:frame])
    if frame > 50:
        sma50_line.set_data(date_indices[49:frame], sma50[49:frame])
    
    # Update buy/sell signals
    buy_points_indices = [df.index.get_loc(date) for date in buy_points if date < dates[frame-1] if frame > 0]
    sell_points_indices = [df.index.get_loc(date) for date in sell_points if date < dates[frame-1] if frame > 0]
    
    if buy_points_indices:
        buy_y = [prices[i] for i in buy_points_indices]
        buy_scatter.set_offsets(np.column_stack([buy_points_indices, buy_y]))
    else:
        buy_scatter.set_offsets(np.zeros((0, 2)))
        
    if sell_points_indices:
        sell_y = [prices[i] for i in sell_points_indices]
        sell_scatter.set_offsets(np.column_stack([sell_points_indices, sell_y]))
    else:
        sell_scatter.set_offsets(np.zeros((0, 2)))
    
    # Update portfolio value
    portfolio_line.set_data(date_indices[:frame], portfolio[:frame])
    
    # Update status text
    if frame > 0:
        in_position = False
        for buy_idx in buy_points_indices:
            for sell_idx in sell_points_indices:
                if buy_idx < frame and (sell_idx > frame or sell_idx < buy_idx):
                    in_position = True
                    break
        
        # If we have a buy signal before the current frame and no sell signal after that buy
        buy_before = any(i < frame for i in buy_points_indices)
        sell_after_last_buy = False
        
        if buy_before:
            last_buy = max(i for i in buy_points_indices if i < frame)
            sell_after_last_buy = any(i > last_buy for i in sell_points_indices if i < frame)
        
        status = "HOLDING STOCK" if (buy_before and not sell_after_last_buy) else "HOLDING CASH"
        value = f"${portfolio[frame-1]:.2f}"
        status_text.set_text(f"Status: {status}\nValue: {value}")
    
    return price_line, sma10_line, sma50_line, portfolio_line, buy_scatter, sell_scatter, status_text

ani = animation.FuncAnimation(fig, update, frames=len(dates), init_func=init, 
                              interval=25, blit=True)

HTML(ani.to_jshtml())
```

```python
from __future__ import (absolute_import, division, print_function, unicode_literals)
import datetime
import os.path
import sys
import backtrader as bt
def get_data(ticker, start, end):
    data = yf.download(ticker, start=start, end=end, multi_level_index=False, interval='1d')
    data.drop(columns = ['Volume'], inplace = True)
    data_feed = bt.feeds.PandasData(dataname=data)
    return data_feed

class TestStrategy(bt.Strategy):
    params = (
        ('maperiod', 15),
    )


    def log(self, txt, dt=None):
        ''' Logging function fot this strategy'''
        dt = dt or self.datas[0].datetime.date(0)
        print('%s, %s' % (dt.isoformat(), txt))

    def __init__(self):
        # Keep a reference to the "close" line in the data[0] dataseries
        self.dataclose = self.datas[0].close

        # To keep track of pending orders and buy price/commission
        self.order = None
        self.buyprice = None
        self.buycomm = None
        self.sma = bt.indicators.MovingAverageSimple(self.datas[0], period=self.params.maperiod)

    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            # Buy/Sell order submitted/accepted to/by broker - Nothing to do
            return

        # Check if an order has been completed
        # Attention: broker could reject order if not enough cash
        if order.status in [order.Completed]:
            if order.isbuy():
                self.log(
                    'BUY EXECUTED, Price: %.2f, Cost: %.2f, Comm %.2f' %
                    (order.executed.price,
                     order.executed.value,
                     order.executed.comm))

                self.buyprice = order.executed.price
                self.buycomm = order.executed.comm
            else:  # Sell
                self.log('SELL EXECUTED, Price: %.2f, Cost: %.2f, Comm %.2f' %
                         (order.executed.price,
                          order.executed.value,
                          order.executed.comm))

            self.bar_executed = len(self)

        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('Order Canceled/Margin/Rejected')

        self.order = None

    def notify_trade(self, trade):
        if not trade.isclosed:
            return

        self.log('OPERATION PROFIT, GROSS %.2f, NET %.2f' %
                 (trade.pnl, trade.pnlcomm))

    def next(self):
        # Simply log the closing price of the series from the reference
        self.log('Close, %.2f' % self.dataclose[0])

        # Check if an order is pending ... if yes, we cannot send a 2nd one
        if self.order:
            return

        # Check if we are in the market
        if not self.position:
            if self.dataclose[0] > self.sma[0]:
                self.log('BUY CREATE, %.2f' % self.dataclose[0])
                self.order = self.buy()

        else:
            if self.dataclose[0] < self.sma[0]:
                self.log('SELL CREATE, %.2f' % self.dataclose[0])
                self.order = self.sell()


cerebro = bt.Cerebro()
cerebro.addstrategy(TestStrategy)
cerebro.broker.setcash(100)
cerebro.broker.setcommission(commission=0.01)
symbol = 'TSLA'
start_date = '2020-03-12'
end_date = '2025-03-11'
data = get_data(symbol, start_date, end_date)
cerebro.adddata(data)
print('Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
cerebro.run()
print('Final Portfolio Value: %.2f' % cerebro.broker.getvalue())
```
