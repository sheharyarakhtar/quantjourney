```python
import yfinance as yf
import pandas as pd
```

```python

def get_data(cryptos):
    df = yf.download(
        tickers = cryptos,
        period = "1mo",
        interval = "15m"
    )
    return df
def clean_crypto_data(df):
    final_df = pd.DataFrame({"Datetime":df.reset_index().Datetime})
    for col in ['Open', 'Close' ,'High', 'Low', 'Volume']:
        temp_df = (df[col]
         .reset_index()
         .melt(id_vars=['Datetime'])
         .rename(columns = {'value':col}))
        final_df = final_df.merge(temp_df, how = 'left')
    return final_df
def create_lagged_features(df, columns = [], lags = []):
    temp = df[columns]
    for col in temp.columns:
        for lag in lags:
            print(lag, col)
            df['lag_'+col+'_'+str(lag)] = df[col].shift(lag)
    return df
```

```python
df = get_data(['BTC-USD','PENDLE-USD'])
final_df = clean_crypto_data(df)
```

```python

create_lagged_features(final_df, columns=['Open','Close'], lags = [2,5,10,15])
```
